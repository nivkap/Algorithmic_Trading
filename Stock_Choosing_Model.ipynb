```python
import pandas as pd
import numpy as np
from pandas_datareader.data import DataReader
import yfinance as yf
from datetime import datetime
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
all_df = pd.DataFrame()
start_date = '2017-01-01'
end_date = '2018-06-01'
fname_string = 'high_beta_stocks.csv'

data = pd.read_csv(fname_string)
tickers = data.Company.values.tolist()

for tkr in tickers:
  single_stock_pd = yf.download(tickers=tkr[0], start=start_date, end=end_date,auto_adjust=True)
  single_stock_pd['Company'] = tkr
  all_df = all_df.append(single_stock_pd)
```

```python
all_df = all_df.reset_index()
all_df['dt'] = pd.to_datetime(all_df['Date'])
all_df['t0'] = (all_df['dt'] - all_df['dt'].min()).dt.days.astype(float)

all_df = pd.merge(all_df, data, on=['Company'], how='left')
all_df
```

```python
def calculate_returns(x):
    return 100*(x.iloc[-1]['Close'] - x.iloc[0]['Open'])/x.iloc[0]['Open']
```

```python
# Split the data to segments (6 months features, 3 months labelling)
FEATURES_SIZE = 180
LABELLING_SIZE = 90

start_train_features_day = 0
end_train_features_day = start_train_features_day + FEATURES_SIZE
start_train_labelling_day = end_train_features_day + 1
end_train_labelling_day = start_train_labelling_day + LABELLING_SIZE
all_features_train_df = all_df[(all_df['t0'] >= start_train_features_day)&(all_df['t0'] < end_train_features_day)]
all_labelling_train_df = all_df[(all_df['t0'] >= start_train_labelling_day)&(all_df['t0'] < end_train_labelling_day)]
```

```python
# Calculating the train returns in the labelling 3 months
train_returns = all_labelling_train_df.groupby('Company').apply(lambda x: calculate_returns(x))
train_returns.hist(bins=100)
```

```python
upper = train_returns.quantile(0.4)
lower = train_returns.quantile(0.2)

good = train_returns[(train_returns > lower) & (train_returns<upper)].index
bad = train_returns[(train_returns < lower) | (train_returns>upper)].index
```

```python
def add_features(df):
  # Creating a monthly df for calculations
  features_df = df.copy()
  features_df['month_index'] = (features_df['t0'] / 30).astype(int)
  monthly_features_df = features_df.groupby(['Company','month_index']).apply(calculate_returns).reset_index(name = 'monthly_return')

  # Min/max price
  min_price = features_df.groupby(['Company','month_index'])['Low'].min().reset_index(name = 'min_price')
  max_price = features_df.groupby(['Company','month_index'])['High'].max().reset_index(name = 'max_price')
  monthly_features_df['min_price'] = min_price['min_price']
  monthly_features_df['max_price'] = max_price['max_price']

  # Average index ratio
  avg_index_returns = monthly_features_df.groupby('month_index')['monthly_return'].mean().reset_index(name = 'avg_index_returns')
  monthly_features_df = monthly_features_df.merge(avg_index_returns, how='inner', left_on='month_index', right_on='month_index')
  monthly_features_df['company_index_ratio'] = monthly_features_df['monthly_return'] / monthly_features_df['avg_index_returns']

  # 6 month return
  all_6_month_return = features_df.groupby('Company').apply(calculate_returns).reset_index(name='6_month_return')

  # Total volume
  total_volume = features_df.groupby(['Company','month_index'])['Volume'].sum().reset_index(name = 'total_volume')
  monthly_features_df['total_volume'] = total_volume['total_volume'] / 1000000

  # Standard deviation
  std = features_df.groupby(['Company'])['Close'].std().reset_index(name = 'std')

  # Average monthly return
  avg_monthly_return = monthly_features_df.groupby(['Company'])['monthly_return'].mean().reset_index(name = 'avg_monthly_return')
  final_df = avg_monthly_return[['Company','avg_monthly_return']]

  # Best/Worst return
  worst_month = monthly_features_df.groupby('Company').min().reset_index()
  best_month = monthly_features_df.groupby('Company').max().reset_index()
  final_df['best_month'] = best_month['monthly_return']
  final_df['worst_month'] = worst_month['monthly_return']

  # min/max price
  highest_price = monthly_features_df.groupby('Company')['max_price'].max().reset_index()
  lowest_price = monthly_features_df.groupby('Company')['min_price'].min().reset_index()
  final_df['max_min_ratio'] = lowest_price['min_price'] / highest_price['max_price']

  # Average index ratio
  average_index_ratio = monthly_features_df.groupby('Company')['company_index_ratio'].mean().reset_index()
  final_df['average_index_ratio'] = average_index_ratio['company_index_ratio']

  # 6 month return
  final_df['6_month_return'] = all_6_month_return['6_month_return']

  # Total volume
  final_df['total_volume'] = total_volume['total_volume']

  # Standard Deviation
  final_df['std'] = std['std']

  return final_df
```

```python
train_df = add_features(all_features_train_df)
train_df
```

```python
# Adding a binary target column for our classification based on the labelling
train_df.loc[train_df['Company'].isin(good),'target'] = 1
train_df['target'].fillna(0,inplace=True)
train_df['target'].value_counts()
```

```python
params = {
    'objective': 'binary:logistic',
    'eval_metric': 'logloss',
    'eta': 0.1,
    'max_depth': 7,
    'scale_pos_weight': 1,
    'random_state': 42
}
```

```python
# Training a XGboost ML model to predict if the stock return will be in the range
X , y =  train_df[['avg_monthly_return','best_month','worst_month', 'max_min_ratio', 'average_index_ratio', '6_month_return', 'total_volume', 'std']] , train_df['target']
clf = XGBClassifier(**params)
clf.fit(X, y)
```

```python
start_test_features_day = 270
end_test_features_day = start_test_features_day + FEATURES_SIZE
start_test_labelling_day = end_test_features_day + 1
end_test_labelling_day = start_test_labelling_day + LABELLING_SIZE

all_features_test_df = all_df[(all_df['t0'] >= start_test_features_day)&(all_df['t0'] < end_test_features_day)]
all_labelling_test_df = all_df[(all_df['t0'] >= start_test_labelling_day)&(all_df['t0'] < end_test_labelling_day)]
```

```python
test_df = add_features(all_features_test_df)
test_df['predicted_y'] = clf.predict(test_df[['avg_monthly_return','best_month','worst_month', 'max_min_ratio', 'average_index_ratio', '6_month_return', 'total_volume', 'std']])
test_df
```

```python
test_df['predicted_y'].value_counts()
```

```python
test_returns = all_labelling_test_df.groupby('Company').apply(lambda x:calculate_returns(x))
upper = train_returns.quantile(0.4)
lower = train_returns.quantile(0.2)

good_test = train_returns[(train_returns > lower) & (train_returns<upper)].index
test_df.loc[test_df['Company'].isin(good_test),'actual_y'] = 1
test_df['actual_y'].fillna(0,inplace=True)
test_df['actual_y'].value_counts()
```

```python
true_positive = len(test_df[(test_df['predicted_y'] == test_df['actual_y'])&(test_df['predicted_y']==True)])
true_negative = len(test_df[(test_df['predicted_y'] == test_df['actual_y'])&(test_df['predicted_y']==False)])
false_positive = len(test_df[(test_df['predicted_y'] != test_df['actual_y'])&(test_df['predicted_y']==True)])
false_negative = len(test_df[(test_df['predicted_y'] != test_df['actual_y'])&(test_df['predicted_y']==False)])
print(
f'''
Train results:
  True positive: {true_positive},
  True negative: {true_negative},
  False positive: {false_positive},
  False negative: {false_negative}
''')
matrix = confusion_matrix(test_df['actual_y'], test_df['predicted_y'])
disp = ConfusionMatrixDisplay(matrix, display_labels=[0,1])
disp.plot()
```
